```python
import numpy as np
from sentence_transformers import SentenceTransformer
from tqdm import tqdm


# Собираем ВСЕ кандидаты
# all_candidates = (
#     legal_noise + technical_essence + stylistic_transform + 
#     grammatical_transform + meta_concepts + patent_constructs
# )

all_candidates = (
    legal_noise 
    
)


# Добавляем противоположные пары как разность векторов
for neg, pos in opposite_pairs:
    emb_neg = model.encode(neg)
    emb_pos = model.encode(pos)
    delta_pair = emb_pos - emb_neg
    similarity = np.dot(delta_avg, delta_pair) / (np.linalg.norm(delta_avg) * np.linalg.norm(delta_pair))
    print(f"Δ ~ ({pos} - {neg}): {similarity:.3f}")

# Проверяем все кандидаты
results = []
for phrase in tqdm(all_candidates):
    emb = model.encode(phrase)
    sim = np.dot(delta_avg, emb) / (np.linalg.norm(delta_avg) * np.linalg.norm(emb))
    results.append((phrase, sim))

# Сортируем и выводим топ-50
results.sort(key=lambda x: abs(x[1]), reverse=True)  # сортируем по модулю!
print("\nТОП-50 кандидатов (по абсолютной близости к Δ):")
for i, (phrase, sim) in enumerate(results[:50], 1):
    sign = "➕" if sim > 0 else "➖"
    print(f"{i:2d}. {sign} {phrase:40s} → {sim:6.3f}")
```
