```python
import pandas as pd
import numpy as np

def compare_search_results(df_a: pd.DataFrame, df_b: pd.DataFrame, id_col: str = 'query_id') -> pd.DataFrame:
    """
    Сравнивает два датафрейма с метриками поиска.
    
    Args:
        df_a: Результаты первого метода (Baseline).
        df_b: Результаты второго метода (Experiment/Delta).
        id_col: Название колонки с ID запроса.
        
    Returns:
        DataFrame со сводной статистикой по каждой метрике.
    """
    
    # 1. Проверка наличия общих ID
    ids_a = set(df_a[id_col])
    ids_b = set(df_b[id_col])
    common_ids = ids_a & ids_b
    
    if len(common_ids) != len(ids_a) or len(common_ids) != len(ids_b):
        print(f"⚠️ Внимание: Не все ID совпадают. Общие: {len(common_ids)}, Только в A: {len(ids_a - ids_b)}, Только в B: {len(ids_b - ids_a)}")
        # Фильтруем только общие, чтобы сравнение было корректным
        df_a = df_a[df_a[id_col].isin(common_ids)].reset_index(drop=True)
        df_b = df_b[df_b[id_col].isin(common_ids)].reset_index(drop=True)
    
    # 2. Определяем список метрик (все колонки кроме ID)
    metrics = [col for col in df_a.columns if col != id_col]
    
    # 3. Вычисляем разницу (Delta = B - A) для каждого семпла
    # Создаем копию, чтобы не менять оригиналы
    diff_df = df_b[metrics].sub(df_a[metrics], axis=0)
    diff_df.insert(0, id_col, df_a[id_col])
    
    # 4. Считаем агрегированную статистику по всем метрикам
    summary_stats = []
    
    for metric in metrics:
        diffs = diff_df[metric].dropna()
        
        mean_diff = diffs.mean()
        std_diff = diffs.std()
        
        # Процент семплов, где метод B лучше метода A
        improved_count = (diffs > 0).sum()
        worsened_count = (diffs < 0).sum()
        unchanged_count = (diffs == 0).sum()
        
        total = len(diffs)
        pct_improved = (improved_count / total * 100) if total > 0 else 0
        pct_worsened = (worsened_count / total * 100) if total > 0 else 0
        
        # Статистическая значимость (парный t-test)
        from scipy import stats
        t_stat, p_value = stats.ttest_rel(df_a[metric].dropna(), df_b[metric].dropna())
        
        summary_stats.append({
            'Metric': metric,
            'Mean_Diff (B-A)': round(mean_diff, 4),
            'Std_Dev': round(std_diff, 4),
            '% Improved': round(pct_improved, 2),
            '% Worsened': round(pct_worsened, 2),
            'P-Value (t-test)': round(p_value, 4),
            'Significant (p<0.05)': '✅ Yes' if p_value < 0.05 else '❌ No'
        })
        
    summary_df = pd.DataFrame(summary_stats)
    
    return summary_df, diff_df
```

```python
import os
import glob
import pandas as pd
import numpy as np
import scipy.stats as stats
# from tabulate import tabulate

def analyze_experiments(
    data_dir: str = ".",
    baseline_file: str = "result_baseline.csv",
    query_id_col: str = "query_id",
    alpha: float = 0.05,
    use_nonparametric: bool = False
):
    baseline_path = os.path.join(data_dir, baseline_file)
    if not os.path.exists(baseline_path):
        raise FileNotFoundError(f"❌ Базовый файл не найден: {baseline_path}")

    # Загружаем baseline
    base_df = pd.read_csv(baseline_path)
    base_df['query_id'] = base_df.index
    if query_id_col not in base_df.columns:
        raise ValueError(f" В baseline файле отсутствует колонка '{query_id_col}'")


    # Определяем метрики (все колонки кроме query_id)
    metrics = [c for c in base_df.columns if c != query_id_col]
    if not metrics:
        raise ValueError("❌ В baseline файле не найдено колонок с метриками")

    # Ищем все файлы result_*.csv кроме baseline
    pattern = os.path.join(data_dir, "result_*.csv")
    files = glob.glob(pattern)
    files = [f for f in files if os.path.basename(f) != baseline_file]
    if not files:
        print("⚠️ Файлы result_*.csv не найдены.")
        return pd.DataFrame()

    results = []

    for fpath in files:
        fname = os.path.basename(fpath)
        # Извлекаем имя метода: result_delta_alpha_minus_3.csv -> delta_alpha_minus_3
        method_name = fname.replace("result_", "").replace(".csv", "")

        exp_df = pd.read_csv(fpath)
        exp_df['query_id'] = exp_df.index
        if query_id_col not in exp_df.columns:
            print(f"⚠️ Пропуск {fname}: нет колонки '{query_id_col}'")
            continue

        # Сопоставляем по query_id
        merged = pd.merge(base_df, exp_df, on=query_id_col, suffixes=("_base", "_exp"))
        
        method_res = {"Метод": method_name, "N_queries": len(merged)}
        
        for m in metrics:
            b_col, e_col = f"{m}_base", f"{m}_exp"
            if b_col not in merged.columns or e_col not in merged.columns:
                continue

            # Убираем пары с NaN
            valid = merged.dropna(subset=[b_col, e_col])
            if len(valid) < 2:
                continue

            # Разница (эксперимент - baseline)
            diff = valid[e_col] - valid[b_col]
            mean_improvement_pp = diff.mean() * 100  # в процентных пунктах

            # Статистический тест
            if use_nonparametric:
                # Wilcoxon signed-rank (непараметрический аналог)
                t_stat, p_two = stats.wilcoxon(valid[e_col], valid[b_col], alternative='greater')
            else:
                # Paired t-test (парный)
                t_stat, p_two = stats.ttest_rel(valid[e_col], valid[b_col])
                # Односторонний p-value: проверяем, что exp > base
                p_val = p_two / 2 if t_stat > 0 else 1.0

            is_sig = p_val < alpha
            method_res[m] = {
                "Δ (п.п.)": round(mean_improvement_pp, 2),
                "p-value": p_val,
                "Значимо": "✅" if is_sig else ""
            }

        results.append(method_res)

    if not results:
        print("⚠️ Не удалось сопоставить ни один файл с baseline.")
        return pd.DataFrame()

    # Преобразуем в удобный DataFrame для вывода
    rows = []
    for r in results:
        row = {"Метод": r["Метод"], "N": r["N_queries"]}
        for m in metrics:
            if m in r:
                stats = r[m]
                row[f"{m} Δ"] = stats["Δ (п.п.)"]
                row[f"{m} p"] = f"{stats['p-value']:.2e}" if stats['p-value'] > 0 else "<1e-16"
                row[f"{m} sig"] = stats["Значимо"]
            else:
                row[f"{m} Δ"] = None
                row[f"{m} p"] = None
                row[f"{m} sig"] = None
        rows.append(row)

    return pd.DataFrame(rows)
```

```python
import pandas as pd
import glob
import os
import re

def aggregate_search_results(folder_path: str = ".") -> pd.DataFrame:
    """
    Агрегирует результаты экспериментов из файлов search_*.csv в единую таблицу.
    """
    # 1. Поиск всех файлов экспериментов
    pattern = os.path.join(folder_path, "result_*.csv")
    files = sorted(glob.glob(pattern))
    
    if not files:
        raise FileNotFoundError(f"❌ Не найдены файлы, начинающиеся с 'result_' в папке: {folder_path}")
        
    summary_data = []
    
    for file_path in files:
        # Извлекаем название эксперимента из имени файла
        file_name = os.path.basename(file_path)
        exp_name = re.sub(r'^result_|\.csv$', '', file_name, flags=re.IGNORECASE)
        
        df = pd.read_csv(file_path)
        
        # Находим колонки с метриками (игнорируем регистр и лишние пробелы)
        metric_cols = [c.strip() for c in df.columns 
                       if re.search(r'(?i)(recall|precision|average_precision|mrr)\s*@', c)]
                       
        # Если файл уже содержит одну строку (предварительно агрегирован), берём как есть
        if len(df) == 1:
            stats = df[metric_cols].iloc[0].to_dict()
        else:
            # Иначе считаем среднее по всем запросам/семплам в файле
            stats = df[metric_cols].mean().to_dict()
            # Опционально: можно добавить std, раскомментировав строку ниже
            # stats.update({f"{k}_std": v for k, v in df[metric_cols].std().items()})
            
        stats['experiment'] = exp_name
        summary_data.append(stats)
        
    # 2. Формируем итоговый DataFrame
    summary_df = pd.DataFrame(summary_data)
    summary_df = summary_df.set_index('experiment')
    
    # 3. Логичная сортировка колонок: сначала по K (5, 10, 20, 30), потом по типу метрики
    metric_order = {'recall': 0, 'precision': 1, 'average_precision': 2, 'mrr': 3}
    
    def sort_key(col_name):
        k_match = re.search(r'(\d+)', col_name)
        k = int(k_match.group(1)) if k_match else 0
        base_name = re.sub(r'(?i)@.*', '', col_name).lower().strip()
        return (k, metric_order.get(base_name, 99), col_name)
        
    sorted_cols = sorted(summary_df.columns, key=sort_key)
    summary_df = summary_df[sorted_cols]
    
    # 4. Округляем для читаемости
    summary_df = summary_df.round(4)
    
    return summary_df

```
